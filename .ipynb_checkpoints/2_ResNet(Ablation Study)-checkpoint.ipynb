## 1. 논문 주요점 살펴보기                 
---

1. 숏컷의 구성
- The shortcut connections simply perform identity mapping, and their outputs are added to the outputs of the stacked layers.       

2. 이 논문이 증명하고자 하는 것        
- 1) Our extremely deep residual nets are easier to optimize, but plain nets exhibit hihger training error when its depth increases.       
- 2) Our residual nets get better accuracy gains than the plain net's.


3. Degradation      
- 연구진은 Degradation problem이 일어난 원인이 identity mapping의 approximation의 어려움에 있을 것이라 가정하여,     
- $H(x) = F(x) + (x)$ 형태로 생각하기 보다는 $F(x) {=} H(x)-x$ 로 보고 identity mapping을 approximation 하는 데 중점을 두었다.      


4. Building Blcok

![빌딩블럭](./PostingPic/2_빌딩블럭.png)        

- In this paper, We consider a building block defined as (upper pic).      
- if the layer is bigger than two,          

> $F = W_2{\sigma}(W_1x)$  즉, F = W1X의 Relued results

__5. Network Architecture__          

##### 1. Plain Network      
- 1. conv layer filter is 3*3 filter.    
- 2. follow 2 design rules:               
   a. 아웃풋 피쳐 맵 사이즈가 같으면, 같은 필터 수를 갖는다.      
   b. 각 레이어의 time complexity를 유지하기 위해, 피쳐 사이즈가 절반이 되면 필터의 수는 2배로 늘어난다.    
- 3. downsampling은 2 strides로, conv 레이어에 바로 적용한다.         
- 4. 네트워크는 global average pooling layer와 1000-way fully-connected layer(with softmax)로 끝난다.      
- 5. 총 레이어의 수는 34개이다. 

##### 2. Residual Network 
- 1. 위의 Plain Network를 베이스로, shortcut connection을 추가한다.      
- 2. Dimension이 증가하면, 두 가지 옵션을 고려한다.       
    a. still performs identity mapping, with extra zero entriespadded for increasing dimentsions.        
    b. match dimension using 1*1 convolution.     
- 3. 2의 옵션을 고려하여, 피쳐맵이 2배로 늘어나면(go across feature maps of two sizes), 2 strides로 수행한다.

6. Implementation        

- 1. imageNet으로 수행     
- 2. 데이터셋에는 random한 scale augmentation을 적용      
- 3. 데이터셋 원본 및 수평 반전(좌우반전) 데이터셋을 224*224 사이즈로 랜덤 크롭       
- 4. standard color augmentation 사용         
- 5. conv layer 이후에 batch normalization -> activation layers        
- 6. minibatch size == 256       

7. 수행 결과 비교       

- 1. 18P(plain) error < 34P error(degradation 문제 발생)      
- 2. 18R(ResNet) error > 34R error         
- 3. 34P error > 34R error    
- 4. 18R converges faster than 18P.            

8. Identity vs Projection Shortcuts        

![비교](./PostingPic/2_comparison.png)        

- 이전의 수행 결과에서, Identity mapping(shortcut)은 훈련 과정에 도움이 됨       
- 추가적으로 Projection shortcut의 효과를 알기 위해 다음의 세 가지 옵션으로 테스트해 봄      
    a. dimension을 확대하고 싶을 때는 zero-padding shortcuts을 적용(모든 shortcut은 parameter free)       
    b. projection shortcut은 dimension을 증가시킬 때만 사용하고, 나머지 부분에는 identity shortcut을 적용       
    c. 모든 shortcut을 projection으로 교체     
    
    
- 결과적으로, 모든 옵션이 일반적인 숏컷(identity)보다 나은 수행 결과를 보임       
- A < B < C 순서로 나은 수행을 보여주었음        
- A<B인 이유 추측 : zero-padding-shortcut일 때 학습되는 부분이 미비해서       
- B<C인 이유 추측 : c의 경우, 전부 projection shorcut이므로 b의 경우보다 학습하는 부분이 많을것임(b는 일부가 identity)     
- __하지만, 결론적으로 A,B,C의 projection shortcuts의 수행 결과로 미루어보아 projection shortcut이 degradation을 해소하는 데 필수 요건은 아님__   



- 따라서 파라미터 및 시간 복잡도를 줄이기 위해 C 옵션은 이번 논문에서 더 이상 사용하지 않으며, Identity shortcut이 다음에 소개할 bottle neck architecture의 복잡도가 증가하는 것을 방지하는 데 중요하다는 사실을 복기한다.

9. Deeper bottleneck Architecture             

- 빌딩 블록을 bottle neck 디자인으로 바꿀 경우         
- 기존의 레이어 디자인을 (1*1 + 3*3 + 1*1 convolution)으로 교체       
- shortcut을 projection shortcuts으로 교체할 경우, 파라미터가 두 배가 되므로, (시간 및 메모리 측면에서) 효율적인 모델을 구성하려면 Identity shortcuts의 존재가 주효함      

__이후의 내용은 구현과 크게 관계없는 결과보고 이므로 생략함__

### 1. ResNet-34                   

![34layer](./PostingPic/2_34layer.jpg)

- 노드에서와 같이, InputShape를 전달하여 summary()할 수 있도록 구성    
- Functional API를 활용하여 빌딩 블록 구성         

__ResNet-34 빌딩블록을 구성하기 위해 고려해야 할 점__       

#### 1. F(x)
1. 커널 사이즈는 3*3으로 고정되어 있으나, 채널 수는 변화가 있음(64,128,256,512...)    
2. 필터 수는 변화가 있음(3,4,6,3...)       
3. 몇 번째 층인지를 알고, 이 층에 따라 내부를 다르게 구성하기 위해 인덱스가 필요할 것임(conv1~conv5)

- conv2~conv5에서 함수 파라미터로 받아야 할 것 : input_shape, kernel_size, filter_count, block_num(블록 개수)   

#### 2. x      
1. 처음에는 데이터셋 인풋과 동일하게,    
2. 이후에는 $x = F(x)$로 지정할 수 있도록 함

#### 3. $F(x) +  x$          
- $F(x)$와 $x$를 더하여 다음 인풋으로 지정

```python
#필요한 라이브러리 임포트 
import tensorflow as tf     
from tensorflow import keras
from tensorflow.python.keras import layers

print("구성에 필요한 라이브러리 임포트 완료")
```

```python
#한 블럭 안에서의 구성 순서는 논문 내용대로   
#conv layer -> batch normalization layer -> activation layer이다.
def building_plain_layer(input_shape=(224,224,3), 
                         kernel_size=(3,3), 
                         block_counts=3, 
                         channel_size=64,
                        layer_index=0):
    
    layer_index=0
    
    #처음으로 들어갈 input_layer를 투입
    x = input_shape
    
    for i in range(block_counts):
        x=keras.layers.Conv2D(filters=channel_size,
                             kernel_size=kernel_size,
                             #관련 요소 논문에 나와있는지 확인
                             padding='same')(x)
        
    #batch normalization-> activation 
    x = keras.layers.BatchNormalization(momentum=0.9)(x)
    x = keras.layers.ReLU()(x)
    
    return x
```

```python
#비교를 위한 34레이어짜리 plain 모델 구성 함수
#input_shape = 첫 구성 함수의 사이즈
#num_classes = 강아지/고양이의 바이너리 밸류이므로 2
def building_plain_34layer(input_shape=(224,224,3),
                          block_counts=[3,4,6,3],
                          channel_list=[64,128,256,512],
                          final_class=2):
    
    assert len(block_counts) == len(channel_list)
    
    #투입할 첫 레이어를 만들어줌
    input_layer = keras.layers.Input(shape=input_shape)
    output=input_layer
    
    #conv1층
    output = keras.layers.Conv2D(filters=64, kernel_size=(7,7), strides=(2,2), name="conv1")(output)
    output = keras.layers.MaxPooling2D(pool_size=(3,3), strides=(2,2), name="conv2_maxpooling")(output)
    
    #conv2층 conv층부터는 위의 함수를 활용하여 수행
    for i, (block, channel) in enumerate(zip(block_counts, channel_list)):
        output = building_plain_layer(output,
                                      kernel_size=(3,3),
                                      block_counts=block,
                                      channel_size=channel,
                                    layer_index=i)
    
    #conv5까지의 conv레이어를 모두 거친 후 처리    
    output = keras.layers.AveragePooling2D(name=f'AveragePooling')(output)
    
    #애버리지 풀링 후 덴스
    output = keras.layers.Dense(final_class, activation='softmax', name='predictions')(output)
    
    model = keras.Model(inputs=input_layer, outputs=output)
    
    return model
```

```python
resnet_34p = building_plain_34layer()
resnet_34p.summary()
```

### 1. 이미지 전처리(사이즈 조정)

### 번외 : 블록 짜는 방법 연습을 위해 VGG논문과 VGG 블록 구성 비교

### 2.1 Architecture                    

1. Input : 224*224 RGB image(preprocessing is subtracting the mean RGB Value)         
2. Through Conv layers(with receptive field 3*3, 1*1 conv filter, convolution strides is fixed to 1 pixel, padding is 3*3) 
      - Spatial pooling is carried out by five maxpooling layers.
      - Maxpooling is performed by 2*2 window, with stride 2.
      
3. After 2, 3FC layers.(이 부분 해석이 모호한데, conv->fcn 인지 fcn->conv인지..)         
      - 1FC, 2FC 는 각각 4096 channel을 가짐
      - 3FC 는 1000개의 channel을 가짐(ILSVRC의 클래스가 1000개 카테고리이기 때문에)      
4. 마지막 레이어는 softmax layer
5. FCN의 경우 모든 블록에서 동일한 configuration을 가짐

```python
def building_vgg_block(input_layer, cnn_count=3, channel=64, block_num=1):
    
    #가장 처음에 제공되는 input_layer를 투입
    x = input_layer
    
    #이후 cnn_count에 따라 반복문을 수행하여 블록 구성
    for count in range(cnn_count):
        
        #여기서는 별도로 strides를 명시하지는 않았음
        #패딩은 3*3으로 명시했으나, 여기서는 same으로 지정하였음
        x = keras.layers.Conv2D(filters=channel, kernel_size=(3,3), activation='relu', kernel_initializer='he_normal',
                               padding='same', name=f'block{block_num}_conv{count}')(x)
        

    #이후에 붙는 maxpooling layers   
    x = keras.layers.MaxPooling2D(pool_size=(2,2), strides=2, name=f'block{block_num}_pooling')(x)
    
    return x
```

- 실제 configuration과 비교       

![vgg](./PostingPic/2_vggnet16.jpg)

```python
#여기서는 input shape가 (32,32,3)
#논문 내용을 적용하여 파라미터만 확인할 것이므로, inputshape를 224,224,3으로 변경
vgg_input_layer = keras.layers.Input(shape=(224,224,3))
vgg_block_output = building_vgg_block(vgg_input_layer)
```

```python
vgg16_model = keras.Model(inputs=vgg_input_layer,outputs=vgg_block_output)

vgg16_model.summary()
```

### VGG-16 통으로 만들기

```python
def building_vgg(input_shape=(224,224,3), 
                 block_counts=[2,2,3,3,3], 
                 channel_list=[64,128,256,512,512], 
                 final_class=10):
    
    #두 configure조건들이 같은 길이인지 확인
    assert len(block_counts) == len(channel_list)
    
    #처음 레이어를 만들어주고
    input_layer = keras.layers.Input(shape=input_shape)
    
    #이제 계속 층을 통과시킬 아웃풋을 input_layer로 초기화함
    output=input_layer
    
    #building block 변수 : input_layer, cnn_count=3, channel=64, block_num=1
    for i, (block, channel) in enumerate(zip(block_counts, channel_list)):
        output = building_vgg_block(output, 
                                   cnn_count=block,
                                   channel=channel,
                                   block_num=i)
        
    
    #블록 구성을 모두 지나고, FCN 3층을 지나
    output=keras.layers.Flatten(name='flatten')(output)
    
    #논문에서 최종 층의 형태를 지정해주었음(1,2FCN은 4096, 마지막 3FCN은 데이터셋 클래스에 맞추어 1000)
    output=keras.layers.Dense(4096,activation='relu', name='fcn1')(output)
    output=keras.layers.Dense(4096,activation='relu', name='fcn2')(output)
    output=keras.layers.Dense(final_class,activation='softmax', name='predictions')(output)
    
    model = keras.Model(inputs=input_layer, outputs=output)
    
    return model    
```

```python
#파라미터 확인을 위해 vgg16을 구성해보자.
#chd vkfkalxj zmrlsms 138millions
vgg_16 = building_vgg()

vgg_16.summary()
```
